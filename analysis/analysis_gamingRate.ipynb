```python
import pandas as pd
import matplotlib.pyplot as plt
# 폰트 오류 해결 코드 ↓
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

FILE_PATH = "./국내게임이용률.xlsx"
df = pd.read_excel(FILE_PATH)
print(df)
```

```python
target_df = df.transpose()
target_df.columns = ['이용률']
target_df.index = target_df.index.astype('str') # 인덱스를 문자열화 시킴
print(target_df)
plt.style.use('ggplot')
plt.plot(target_df.index, target_df.values, 'ro--', markersize=10)
plt.ylim(50, 80)
plt.title("국내 게임 이용률")
```
