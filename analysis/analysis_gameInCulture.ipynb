```python
import pandas as pd
import matplotlib.pyplot as plt
# 폰트 오류 해결 코드 ↓
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

FILE_PATH = "./문화콘텐츠동향.xlsx"
df = pd.read_excel(FILE_PATH)
df.set_index('문화콘텐츠',inplace=True)
print(df)
```

```python
plt.style.use('ggplot')
fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(1, 1, 1)

ax.plot(df.columns, df.loc['음악'], marker='o', markerfacecolor='green', 
        markersize=10, color='olive', linewidth=2, label='음악')
ax.plot(df.columns, df.loc['게임'], marker='o', markerfacecolor='pink', 
        markersize=15, color='red', linewidth=3, label='게임')
ax.plot(df.columns, df.loc['영화'], marker='o', markerfacecolor='darkblue', 
        markersize=10, color='skyblue', linewidth=2, label='영화')
ax.plot(df.columns, df.loc['방송'], marker='o', markerfacecolor='darkmagenta', 
        markersize=10, color='purple', linewidth=2, label='방송')
ax.plot(df.columns, df.loc['광고'], marker='o', markerfacecolor='gray', 
        markersize=10, color='dimgray', linewidth=2, label='광고')

ax.legend(loc='best')
plt.ylim(1500, 10000)
```
