```python
import pandas as pd
import matplotlib.pyplot as plt
# 폰트 오류 해결 코드 ↓
from matplotlib import font_manager, rc
font_path = "./malgun.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

FILE_PATH = "./외국게임산업동향.xlsx"
df = pd.read_excel(FILE_PATH)
df.set_index('국가',inplace=True)
print(df)
```

```python
df.columns = df.columns.astype('str') # 열을 문자열화 시킴
plt.style.use('ggplot')

fig = plt.figure(figsize=(15, 10))
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)

ax1.plot(df.columns, df.loc['미국'], marker='o', markerfacecolor='khaki', 
        markersize=10, color='gold', linewidth=2, label='미국')
ax2.plot(df.columns, df.loc['중국'], marker='o', markerfacecolor='coral', 
        markersize=10, color='lightsalmon', linewidth=2, label='중국')
ax3.plot(df.columns, df.loc['일본'], marker='o', markerfacecolor='deeppink', 
        markersize=10, color='hotpink', linewidth=2, label='일본')
ax4.plot(df.columns, df.loc['러시아'], marker='o', markerfacecolor='turquoise', 
        markersize=10, color='skyblue', linewidth=2, label='러시아')

ax1.legend(loc='best')
ax2.legend(loc='best')
ax3.legend(loc='best')
ax4.legend(loc='best')
```
